### User Defined Functions


```python
def PrintLoanData(dataframe, by, loanstatus='Charged Off'):
    grp = dataframe.groupby(['loan_status',by])[by].count()
    cnt = dataframe.groupby(by)[by].count()
    percentages = grp.unstack() * 100 / cnt.T
    return percentages
```

```python
def drawLoanGraph(dataframe, by, loanstatus='Charged Off'):
    grp = dataframe.groupby(['loan_status',by])[by].count()
    cnt = dataframe.groupby(by)[by].count()
    #print(grp)
    percentages = grp.unstack() * 100 / cnt #grp.unstack()
    #print(percentages)
    ax = percentages.loc[loanstatus].plot.bar(color=sns.color_palette('husl', 16))
    ax.set_ylabel('% of loans ' + loanstatus)
    annotate_df(ax, grp[loanstatus].to_dict())
    plt.margins(0.2, 0.2)
    plt.tight_layout()
    return ax
```

```python
def annotate_df(ax, feature):
    total = len(feature)
    for p in ax.patches:
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        percentage = "{:.1f}%".format(y)
        
        ax.annotate(percentage, (x, y), size = 12)
```

```python
def without_hue(plot, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
    plt.show()
```

5. `Home Ownership and Annual Income on Loan Status:`

```python
# Get annual income and home ownership in a seperate dataframe
data_annual_inc_ls = data_chargedoff.loc[:,['annual_inc','home_ownership']]

# Add the income level labels to the dataframe and get the count of charged off in it
data_annual_inc_ls['inc_level'] = pd.cut(data_annual_inc_ls.annual_inc, bins = annualinc_bins)
data_annual_inc_ls = data_annual_inc_ls.loc[:, ['inc_level', 'home_ownership']]
data_annual_inc_ls = data_annual_inc_ls.groupby(['inc_level', 'home_ownership']).size()
data_annual_inc_ls = data_annual_inc_ls.reset_index()
data_annual_inc_ls.rename({0:'Charged Off'}, axis='columns', inplace = True)

# Convert the count into percentage of charged off cases in each home ownership for a given income level
# Get the sum in each income level
choff_inc_cnt = data_annual_inc_ls.groupby('inc_level').sum().reset_index()
# Merge the sum in each income level with the count in each home ownership status
data_annual_inc_ls = pd.merge(data_annual_inc_ls, choff_inc_cnt, how = 'inner', on = 'inc_level')
# Now get the percentage of charged off cases in each home ownership for a given income level
data_annual_inc_ls['Charged Off_x'] = round(100 * data_annual_inc_ls['Charged Off_x'] / data_annual_inc_ls['Charged Off_y'] , 2)
data_annual_inc_ls = data_annual_inc_ls.drop('Charged Off_y', axis=1)
data_annual_inc_ls.rename({'Charged Off_x':'Charged Off'}, axis='columns', inplace = True)
data_annual_inc_ls.style.format({'Charged Off':'{0:,.2f}%'})
```

```python
# Plot the above table in a chart

# set the seaborn theme
sns.set(style = "whitegrid")
# adjust figure size 
plt.figure(figsize=(16,6))

# Draw the Plot
ax = sns.barplot(y='Charged Off', x='inc_level', hue = 'home_ownership', data= data_annual_inc_ls)
# Get values of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center") 
    
plt.xlabel("Income Level")
plt.ylabel("Percentage of Defaulters")
plt.legend(title = "Home Ownership")
plt.xticks(rotation=45)
plt.show()
```

<b> Observation</b>
<ul><li>Less Income with rented housetend to default the loan more. </li>
<li>Less Income having own house are also tend to defaulted but lesser than the rented one. </li>
<li>With High Income the borrowers living in Mortgaged houses tend to default the loan more. </li>
    </ul>
